```python
import os
import sys
import urllib.request
import ssl
import pandas as pd
import json

client_id = "VVrSb7DXLYUr2IRlQNSU"
client_secret = "pJSQFrP13Z"
encText = urllib.parse.quote("코로나")

display_num = 10

result_list = [] #마지막 10개만 들어가지 않게 하기 위해서 리스트형태로 설정
for start in range(1, 201, 10): #중간에 갯수를 설정

    url = "https://openapi.naver.com/v1/search/news?query=" + encText + "&display=" + str(display_num) + "&start=" + str(start) 
    #결과를 더 가져오기 위해서 파라미터를 수정함
    # url = "https://openapi.naver.com/v1/search/news?query=" + encText # JSON 결과
    # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    context = ssl._create_unverified_context()
    response = urllib.request.urlopen(request, context=context)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        result = response_body.decode('utf-8') #이 부분도 살짝 변형함 result 딕셔너리로 선언
        json_result = json.loads(result)
        result_list = result_list + (json_result['items'])
    else:
        print("Error Code:" + rescode)
```

```python
import konlpy
from konlpy.tag import Okt #한국어 품사구분을 해주는 라이브러리
import re

jvm_path = "/Library/Java/JavaVirtualMachines/zulu-15.jdk/Contents/Home/bin/java"
okt = Okt(jvmpath=jvm_path)
```

```python
all_news_title = ""
for new_title in title_list:
    all_news_title = all_news_title + new_title  #하나의 문장으로 합쳐짐
print(all_news_title)
```

```python
nlp = okt
#nlp = Okt() #원래는 이 코드인데 패스를 따로 설정해서 위처럼 따로 변수에 저장함
message_N =nlp.nouns(all_news_title) #명사만 거르게 됨
print(message_N)
```

```python
from collections import Counter
count = Counter(message_N) #메세지가 몇개가 나왔는지 카운트
tags = count.most_common(50) #모스트 50 빈도 수가 가장 많이 나온
print(tags)
```
