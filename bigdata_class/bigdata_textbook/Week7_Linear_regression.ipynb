```python
import pandas as pd
from sklearn import datasets
```

```python
red_df = pd.read_csv("./winequality-red.csv", sep=";") # ";"을 기준으로 컬럼을 구분하여 csv파일을 읽고 변수에 저장
white_df = pd.read_csv("./winequality-white.csv", sep=";")
```

```python
print (red_df.head())
```

```python
print (white_df.head())
```

```python
red_df.insert(0, column='type', value='red') #타입을 레드 기준으로 정렬
white_df.insert(0, column='type', value='white')
```

```python
print (red_df)
```

```python
print (white_df)
```

## red wine, white wine dataframe 을 합쳐서 하나의 데이터프레임으로 만들어라.

```python
wine = pd.concat([red_df, white_df], ignore_index=True)
print (wine)
```

```python
print (wine.info())
```

```python
wine.describe()
```

```python
wine.quality.unique()
```

```python
wine.quality.value_counts()
```

```python
wine.groupby('type')['quality'].describe()
```

```python
wine['type'] = wine['type'].replace('red','1')
wine['type'] = wine['type'].replace('white','0')
```

```python
print (wine)
```

```python
X = wine.drop(['quality'], axis=1)
y = wine['quality']
```

```python
print (X)
```

```python
print (y)
```

```python
from sklearn.model_selection import train_test_split

### Train : Test = 80 : 20
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) 
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print (y_pred)
```

## MAE (Mean Absolute Error), MSE (Mean Squared Error)




```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
print (mae)

mse = mean_squared_error(y_test, y_pred)
print (mse)
```
