```python
import requests # HTTP 요청을 처리하는 모듈로, 웹 서버로부터 데이터를 가져오거나 웹 서버에 데이터를 전송하기 위해 사용
from bs4 import BeautifulSoup # Python에서 웹 스크래핑을 위해 사용되는 라이브러리인 BeautifulSoup을 임포트하는 것
#BeautifulSoup은 HTML, XML 등의 웹 페이지의 구조를 파싱하고, 원하는 데이터를 추출하는데 사용
#bs4는 BeautifulSoup 모듈이 포함된 beautifulsoup4 라이브러리를 가져온다는 의미
```

```python
url = "https://www.yelp.com/search?find_desc=&find_loc=New+York%2C+NY&start=0" #url를 하나 지정
response = requests.get(url) # HTML 코드를 가져온다. request.get(url)을 사용해서 
#웹 서버에 GET 방식으로 요청을 보내고, 해당 URL의 웹 페이지의 내용을 가져올 수 있음
```

```python
print (response.text) #.text를 사용해서 text 형식으로 가져옴 붙이지 않으면 통신이 된다는 response = 200 만 표시
```

```python
soup = BeautifulSoup(response.text, 'html.parser') # HTML 코드를 HTML 형식으로 구문 분석한다.HTML 문서의 구조를 파싱하여 원하는 요소를 쉽게 추출하고 처리
#텍스트 구문으로 파싱하는게 좋음
print (soup)
```

```python
restaurant_title_results = soup.find_all('a', {'class': 'css-1m051bw'}) # HTML 코드에서 Class가 "css-1m051bw"를 가지는 a 태그는 주요로 링크 요소
#'find_all()' 메서드는 BeautifulSoup로 조건에 맞는 모든 요소를 찾아서 리스트로 반환합니다. 이때 첫 번째 인자로는 찾고자 하는 HTML a 태그를 지정,
#두 번째 인자로는 해당 HTML 태그의 속성과 값 'css-1m051bw' 을 딕셔너리 키 : 벨류 형태로 지정
for result in restaurant_title_results: # 모든 태그들의 내용을 for문으로 검출
    print(result.text) #태그의 내용(content)에 해당하는 텍스트 값을 가지고 있습니다. '.text' 속성을 사용하여 해당 텍스트 값을 추출
```

```python
review_count = soup.find_all('span', {'class': 'css-chan6m'}) #리뷰의 총 갯수를 현재 페이지에서 모두 찾음
for result in review_count:
    print (result.text) # 텍스트로 표현
#그런데 이방법은 쓸모 없는 정크데이터가 너무 많음
```

## Yelp 의 New York에 있는 인기식당의 1. 이름과 2. 어떤 음식 종류를 파는지 1페이지 검색 결과를 출력해보세요.

```python
url_list = list() #리스트를 만들고
for number in [0, 10, 20, 30, 40]: #리스트의 갯수만큼 반복
    url_list.append("https://www.yelp.com/search?find_desc=&find_loc=New+York%2C+NY%2C+United+States&start=" + str(number)) #링크 페이지를
    #넘기기 위해 리스트를 추가 str(number)로 바꿔주는 것이 중요

for url in url_list: #링크페이지가 제대로 넘어가는지 프린트해서 확인
    print (url)
```

```python
url = url_list[0] #페이지 첫번째를 url 변수에 넣음
response = requests.get(url)# request.get(url)을 사용해서 
#웹 서버에 GET 방식으로 요청을 보내고, 해당 URL의 웹 페이지의 내용을 가져올 수 있음
soup = BeautifulSoup(response.text, 'html.parser') # HTML 코드를 HTML 형식으로 구문 분석(파싱)한다. response.text로 파싱하는 것이 중요
restaurant_list = soup.find_all('div', {'data-testid':'serp-ia-card'}) #하나의 박스데이터를 모두 가져옴
for rest in restaurant_list: #전부 순회하면서 rest에 집어넣음
    print (rest) #하나의 박스 html을 모두 가져옴
```

```python
for rest in restaurant_list: # restaurant_title_results = soup.find_all('a', {'class': 'css-1m051bw'}) 박스데이터를 rest에 전부 집어넣음
    name = rest.find_all('a', {'class': 'css-1m051bw'})[0] #제목데이터와 정크데이터가 있는 태그에서 제목데이터만 반환하기 위해[0] 선언
    restaurant_name = name.text #제목을 텍스트로 변환
    
    type_list = rest.find_all('p', {'class': 'css-dzq7l1'})[0] #해쉬태크가 연속 3개 있는 데이터 박스의 데이터에서 첫번째 이름 데이터만 가져옴
    #print(type_list) <p class="css-dzq7l1"><span class="display--inline__09f24__c6N_k border-color--default__09f24__NPAKY"> 이런 값도 가져옴
    
    restaurant_type = list() #리스트를 하나 추가
    for type in type_list: #정크 값을 없애기 위해 해쉬태그 이름의 갯수만큼 순회
        each_type = type.find_all('span', {'class':'css-11bijt4'}) #한번 더 이름만 있는 태그로 데이터를 거름 이름의 요소를 선택
        
        if len(each_type) >= 1: # 기본적으로 'css-11bijt4'의 이름 태그를 3개씩 가지고 있으므로 태그 이름이 1개보다 크면
            for type2 in each_type: #타입 개수 만큼 순회
                restaurant_type.append(type2.text) #리스트에 텍스트 형식으로 추가
        # for ele in each_type.find_all('a', {'class':'css-q7yb35'}):
        #     print (ele)'
        
    types = ','.join(restaurant_type) 
    # 리스트의 요소들을 쉼표(,)로 구분하여 문자열로 합치는 역할, 'restaurant_type' 리스트는 이전에 해당 레스토랑의 해시태그 데이터를 담아둔 리스트
    # 각 해시태그들을 문자열로 합치기 위해 .join() 사용 'restaurant_type' 리스트에 있는 해시태그들을 쉼표(,)로 구분하여 하나의 문자열로 만듬

    print ("Restaurant Name: " + restaurant_name + "\t\t\ttypes: " + types) #\t\t\t 탭 3개의 간격
    
```

## Yelp 의 New York에 있는 인기식당의 1. 이름과 2. 어떤 음식 종류를 파는지 1페이지 부터 5페이지까지의 검색 결과를 출력해보세요

```python
url_list = list() #리스트 선언
for number in [0, 10, 20 , 30]: # 페이지 링크를 넘기기 위한 리스트 선언
    url_list.append("https://www.yelp.com/search?find_desc=&find_loc=New+York%2C+NY%2C+United+States&sortby=review_count&start=" + str(number)) #str(number)로 만들어 주는 것이 주요
#이 작업으로 인해 온전한 페이지를 넘길 수 있는 페이지 url이 담긴 리스트를 만듬 
import time # 타임슬립을 하기 위한 임포트


for url in url_list: # url 4개를 손해
    response = requests.get(url) # url의 get을 얻어서 response에 저장
    soup = BeautifulSoup(response.text, 'html.parser') # HTML 코드를 HTML 형식으로 구문으로 파싱(분석) response.text로 하는 것 중요!
    #BeautifulSoup(response.text, 'html.parser') 정해진 형식 중요
    time.sleep(10) #페이지로딩을 위해 타임 슬립을 10초를 줌 
    
    restaurant_list = soup.find_all('div', {'data-testid':'serp-ia-card'}) # 음식점 박스데이터를 가져옴
    for rest in restaurant_list:#박스 데이터 만큼 순회
        name = rest.find_all('a', {'class': 'css-1m051bw'})[0] #제목 데이터를 가져옴, 첫번째 요소를 반환
        restaurant_name = name.text # 텍스트로 만듬
        
        type_list = rest.find_all('p', {'class': 'css-dzq7l1'})[0] #해쉬태그가 있는 박스데이터를 가져옴, 첫번째 요소를 반환
        
        restaurant_type = list() # 리스트를 생성
        for type in type_list: # 해쉬태그를 순회
            each_type = type.find_all('span', {'class':'css-11bijt4'}) #정확한 데이터 추출을 위해 해쉬태그를 한번더 구분지음
            
            if len(each_type) >= 1: # 추출된 해쉬 태그 이름이 1개 이상일 때 1개 이하일때는 없으므로,
                for type2 in each_type: # 해쉬태그 갯수만큼 반환
                    restaurant_type.append(type2.text) #리스트에 텍스트 형태로 추가
            # for ele in each_type.find_all('a', {'class':'css-q7yb35'}):
            #     print (ele)'
            
        types = ','.join(restaurant_type) #.join함수를 써서 restaurant_type 리스트에 들어 있는 이름 데이터를 ,로 구분지음

        print ("Restaurant Name: " + restaurant_name + "\t\t\ttypes: " + types) #프린트 \t\t tap 2번 만큼의 스페이스
```

```python
from selenium.webdriver.common.by import By
images = (driver.find_elements(By.XPATH, '//img[@class="rg_i Q4LuWd"]')) # 그림 이미지의 클래스를 지정
#By.XPATH를 사용하여 XPath를 이용하여 엘리먼트를 찾고, //img[@class="rg_i Q4LuWd"]는 이미지 엘리먼트를 찾기 위한 XPath 경로를 나타냄
# 이 경로는 웹 페이지에서 class 속성이 "rg_i Q4LuWd"인 img 엘리먼트를 찾는 것을 의미
# 딕셔너리로 지정할 수 없음
```

```python
from io import BytesIO
#from io import BytesIO는 파이썬의 io 모듈에서 BytesIO 클래스를 임포트하는 것,
# BytesIO 클래스는 메모리에 바이너리 데이터를 버퍼로 저장하는데 사용되며, 파일과 같이 다룰 수 있음
# 이를 통해 메모리에서 바이너리 데이터를 읽고 쓸 수 있습
from PIL import Image # 필로우를 임포트
for i, image in enumerate(images):#애너머레이트로 인덱스 값과 리스트 형태 이미지
    image_url = image.get_attribute("src") #
    if image_url:
        try:
            response = requests.get(image_url)
            image_content = response.content

            image_pil = Image.open(BytesIO(image_content))

            file_name = f"dog_{i}.jpg"
            with open(file_name, "wb") as f:
                image_pil.save(f, "JPEG")
                print(f"Saved {file_name}")
        except:
            continue
```
