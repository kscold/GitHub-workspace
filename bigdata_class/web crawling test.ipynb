```python
import requests
from bs4 import BeautifulSoup
url = "https://www.yelp.com/search?find_desc=&find_loc=New+York%2C+NY%2C+United+States"#url을 변수에 저장
response = requests.get(url)#url을 가져옴 변수에 이 코드를 너무 많이 쓰면 리퀘스트가 ip기반으로 못하게 할 수도 있음
print(response.text) # 200이 인터넷에서 OK이라는 뜻 .text을 붙여서 내용을 표현

soup = BeautifulSoup(response.text, 'html.parser')#내부적으로 바뀜
print(soup)
```

```python
restaurant_title_result = soup.find_all('a',{'class': 'css-1m051bw'})#a라는 이름에 HTML 코드에서 베이커리 클래스를 복사
for result in restaurant_title_result: # 결과를 result로 넣음
    print(result.text)#타이틀 이름만 저장
```

```python
restaurant_count_review = soup.find_all('span',{'class': 'css-chan6m'})#HTML 코드에서 베이커리 클래스를 복사
for review_count in restaurant_count_review: # 결과의 갯수반큼 프린트하기 위해
    print(review_count.text)#타이틀 이름만 저장 그런데 span이라는 값을 가지고 있는 값들이 많기 때문에 추가적인 필터링이 필요
    ##print("\n\n") 홀수번만 가져오는 코드
    review_count2 = soup.find_all('div', {'data-testid':'serp-ia-card'})#쓸모 없는 값을 거르기 위해 박스 안에 div에서 검출
    for review in review_count2:
        review_text = review.find_all('span',{'class': 'css-chan6m'})[0] #리뷰카운트에서 첫번째 것만
        print(review_text.text)
```

```python
url = "https://www.google.co.kr/search?q=cat&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjxsYCvjfH9AhUSXmwGHXQoCMAQ_AUoAXoECAEQAw&biw=1440&bih=764&dpr=2"
driver.get(url)
```

```python
from selenium.webdriver.common.by import By
images = (driver.find_elements(By.XPATH, '//img[@class="rg_i Q4LuWd"]'))
```

```python
from io import BytesIO
from PIL import Image
for i, image in enumerate(images):
    image_url = image.get_attribute("src")
    if image_url:
        try:
            response = requests.get(image_url)
            image_content = response.content

            image_pil = Image.open(BytesIO(image_content))

            file_name = f"dog_{i}.jpg"
            with open(file_name, "wb") as f:
                image_pil.save(f, "JPEG")
                print(f"Saved {file_name}")
        except:
            continue
```
