```python
import pandas as pd

#1번 문제 코드
read_data_bata = pd.read_csv("exams_long.csv", sep=",")#초기 데이터를 가져옴

read_data = read_data_bata.drop(['lunch', 'parental level of education'], axis=1)#필요없는 열 삭제

print(read_data)#데이터 출력
print("")

#2번 문제 코드
print("math score의 평군 점수는 = ",read_data['math score'].mean())#읽어온 데이터의 수학 열 점수 평균 출력
print("reading score의 평군 점수는 = ",read_data['reading score'].mean())#읽어온 데이터의 읽기 열 점수 평균 출력
print("writing score의 평군 점수는 = ",read_data["writing score"].mean())#읽어온 데이터의 쓰기 열 점수 평균 출력
print("")

#3번 문제 코드
group_A = read_data.loc[read_data['race/ethnicity'] == 'group A']#race/ethnicity 열에서 주어진 조건의 행(그룹 A의 것만) 검출
group_B = read_data.loc[read_data['race/ethnicity'] == 'group B']#race/ethnicity 열에서 주어진 조건의 행(그룹 B의 것만) 검출
group_C = read_data.loc[read_data['race/ethnicity'] == 'group C']#race/ethnicity 열에서 주어진 조건의 행(그룹 C의 것만) 검출


group_A_sub = group_A[['math score', 'reading score', 'writing score']].mean()#그룹 A 중 수학,읽기,쓰기(타이틀) 평균점수를 저장
group_B_sub = group_B[['math score', 'reading score', 'writing score']].mean()#그룹 B 중 수학,읽기,쓰기(타이틀) 평균점수를 저장
group_C_sub = group_C[['math score', 'reading score', 'writing score']].mean()#그룹 C 중 수학,읽기,쓰기(타이틀) 평균점수를 저장

group_all_mean = pd.DataFrame([group_A_sub, group_B_sub, group_C_sub], index=['group A', 'group B', 'group C'], columns=['math score', 'reading score', 'writing score'])# 그룹 A,B,C 열, 수학,읽기,쓰기 행으로 데이터프레임
print("group별 평균 점수")
print(group_all_mean)
print("")

#4번 문제 코드
group_C_male = group_C[group_C['gender'] == 'male']#그룹 C 안에 성별 열에서 남성인 행데이터들만 검출
group_C_male_math_mean = group_C_male['math score'].mean()#수학 점수 열의 평균값을 변수에 저장
group_C_male_reading_mean = group_C_male['reading score'].mean()#읽기 점수 열의 평균값을 변수에 저장
group_C_male_writing_mean = group_C_male['writing score'].mean()#쓰기 점수 열의 평균값을 변수에 저장
group_C_male_all_mean = pd.DataFrame({'math score': [group_C_male_math_mean], 'reading score': [group_C_male_reading_mean], 'writing score': [group_C_male_writing_mean]}, index=['group C male'])#수학,읽기,쓰기 열에 평균 점수을 넣고 행타이틀을 group C male로 데이터 프레임을 만듬
print("group C의 남자 평균 점수")
print(group_C_male_all_mean)
print("")

group_C_female = group_C[group_C['gender'] == 'female']#그룹 C 안에 성별 열에서 여성인 행데이터들만 검출
group_C_female_math_mean = group_C_female['math score'].mean()#수학 점수 열의 평균값을 변수에 저장
group_C_female_reading_mean = group_C_female['reading score'].mean()#일기 점수 열의 평균값을 변수에 저장
group_C_female_writing_mean = group_C_female['writing score'].mean()#쓰기 점수 열의 평균값을 변수에 저장
group_C_female_all_mean = pd.DataFrame({'math score': [group_C_female_math_mean], 'reading score': [group_C_female_reading_mean], 'writing score': [group_C_female_writing_mean]}, index=['group C female'])#수학,읽기,쓰기 열에 평균 점수을 넣고 행타이틀을 group C female로 데이터 프레임을 만듬
print("group C의 여자 평균 점수")
print(group_C_female_all_mean)
print("")

#5번 문제
male_math_80 = read_data[(read_data['gender'] == 'male') & (read_data['math score'] >= 80)]#성별 열에서 남성인 경우와 수학열의 데이커가 80점 이상인 경우을 동시에 만족하는 데이터 프레임을 검출
num_male_math_80 = len(male_math_80)#이 데이터프레임의 행 갯수를 변수에 저장
print("수학 점수가 80점 이상인 남자 학생들의 수:", num_male_math_80)
print("")

#6번 문제
remain_completed = read_data[read_data['test preparation course'] == 'completed']#시험 준비 열에서 준비를 완료한 학생들만 검출
print("시험 준비 과정을 완료 한 학생들")
print(remain_completed)
print("")

#7번 문제
import random #랜덤함수를 사용하기 위한 라이브러리 가져옴
remain_completed = remain_completed.drop(['test preparation course'], axis=1)#앞서 시험에 통과한 학생들 밖에 없으므로 시험준비 열을 삭제해서 데이터 줄이기
listening_score = [random.randint(0, 100) for _ in range(len(remain_completed))]#남은 학생들의 행데이터 갯수 반큼 랜덤 난수 점수를 생성
remain_completed.loc[:, 'listening score'] = listening_score#듣기 점수 열 생성과 열데이터 추가
print("시험 준비 과정을 완료 후 ristening score가 추가된 학생들")
print(remain_completed)
```
