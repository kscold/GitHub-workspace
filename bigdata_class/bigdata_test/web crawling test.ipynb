```python
import requests
from bs4 import BeautifulSoup
url = "https://www.yelp.com/search?find_desc=&find_loc=New+York%2C+NY%2C+United+States"#url  
response = requests.get(url)#url         ip    
print(response.text) # 200  OK  .text   

soup = BeautifulSoup(response.text, 'html.parser')# 
print(soup)
```

```python
restaurant_title_result = soup.find_all('a',{'class': 'css-1m051bw'})#a  HTML    
for result in restaurant_title_result: #  result 
    print(result.text)#  
```

```python
restaurant_count_review = soup.find_all('span',{'class': 'css-chan6m'})#HTML    
for review_count in restaurant_count_review: #    
    print(review_count.text)#    span         
    ##print("\n\n")   
    review_count2 = soup.find_all('div', {'data-testid':'serp-ia-card'})#       div 
    for review in review_count2:
        review_text = review.find_all('span',{'class': 'css-chan6m'})[0] #  
        print(review_text.text)
```

```python
from selenium.webdriver.common.by import By
images = (driver.find_elements(By.XPATH, '//img[@class="rg_i Q4LuWd"]'))
```

```python
from io import BytesIO
from PIL import Image
for i, image in enumerate(images):
    image_url = image.get_attribute("src")
    if image_url:
        try:
            response = requests.get(image_url)
            image_content = response.content

            image_pil = Image.open(BytesIO(image_content))

            file_name = f"dog_{i}.jpg"
            with open(file_name, "wb") as f:
                image_pil.save(f, "JPEG")
                print(f"Saved {file_name}")
        except:
            continue
```
