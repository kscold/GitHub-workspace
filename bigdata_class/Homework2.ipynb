```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url_list = []#url 리스트 생성
for page in range(1, 6):#1페이지부터 6페이지까지
    url = f'https://www.genie.co.kr/genre/M0100?genreCode=M0100&pg={page}' # HTML 코드를 HTML 형식으로 구문 분석한다.
    url_list.append(url)#리스트에 추가
    print(url_list)#프린트 하여 url이 리스트에 잘 들어갔는지 확인
```

```python
songs = []#음악 데이터를 저장할 리스트 만듬
for url in url_list:#5개 url을 반복
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}#보안으로 인한 헤더 적용(맥용 크롬 기준)

    response = requests.get(url, headers=headers)#해더 파일 설정 밑 html코드 가져오기
    soup = BeautifulSoup(response.text, 'html.parser')# html 코드를 html 형식으로 구문 분석
    titles = soup.find_all('td', {'class': 'info'})#박스정보데이터를 추출


    for title in titles:#추출한 갯수만큼 반복
        song = title.find_all('a',{'class': 'title ellipsis'})[0].text#제목을 찾고 첫번째 요소를 텍스트로 반환
        artist = title.find_all('a',{'class': 'artist ellipsis'})[0].text#가수를 찾고 첫번째 요소를 텍스트로 반환
        album = title.find_all('a',{'class': 'albumtitle ellipsis'})[0].text#앨범을 찾고 첫번째 요소를 텍스트로 반환
        songs.append({'제목': song, '가수': artist, '앨범': album})#리스트에 정보를 추가
        print(f"{song} - {artist} ({album})")#결과 프린트

song_data = pd.DataFrame(songs)#데이터 프레임으로 변환
song_data.index += 1#csv인덱스를 1부터 바꿈
song_data.to_csv("./song_chart.csv")#음악 데이터 파일을 csv로 저장
```
